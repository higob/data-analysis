# 4.Correlation analysis of dataset

# Correlation modeling
Correlation is a statistical term that describes the process of determining the links between two variables (BYJUS, 2021).
I have used pandas to generate a table showing a correlation between all the factors. I took a variable corr and used it. corr () function and then print the variable to get the table showing a correlation between all the numerical columns of the dataset.

```python
#Let's plot a correlation graph to find out correlation of various factors with each other.
corr= property.corr()
corr
```

                                              1.3 Table of correlation

To plot the correlation figure, I used matplotlib. I specified the figure size and used plt.matshow function in which I specified the corr variable , color scheme, and then fignum. I used plt.xticks to specify the x-axis and plt.yticks to specify the y-axis.

```python
#Let's visualize correlation for better understanding.
fig = plt.figure(figsize=(8,8))
plt.matshow(corr,cmap='RdBu',fignum=fig.number)
plt.xticks(range(len(corr.columns)),corr.columns,rotation='vertical');
plt.yticks(range(len(corr.columns)),corr.columns);
```
